```python
# Drop the columns
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
#  drop the columns
df.drop(['car_name', 'brand'],axis=1,inplace=True)
```

```python
# Seen the numerical and categorical data
numerical_ferature=[feature for feature in df.columns if df[feature].dtype!="object"]
categorical_feature=[feature for feature in df.columns if df[feature].dtype=='object']
print("categorical_feature",len(categorical_feature))
print('numerical_ferature',len(numerical_ferature))
```

Split the data

```python
#  split the data into dependent event aur independent event
X=df.drop(['selling_price'],axis=1)
y=df['selling_price']
```

```python
# Label encoding
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
X["model"]=label.fit_transform(X["model"])
```

```python
# Train to split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_train.shape,X_test.shape
```

Transform the catgorical columns

```python
df.head()
```

```python
# use onehotencoding labelencoding and standardscaler
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer


cat_feature=['seller_type','fuel_type','transmission_type']
num_feature=X.select_dtypes(exclude=object).columns

# create the object of the  encoder
one_hot_encoder=OneHotEncoder(drop='first')
Standard_Scaler=StandardScaler()


Transform=ColumnTransformer([
    ("Onehotencoding",one_hot_encoder,cat_feature),
    ("Standardscaler",Standard_Scaler,num_feature)
],remainder="passthrough")
```

Transform the coloums


```python
# encoding perform
X_train=Transform.fit_transform(X_train)
X_test=Transform.transform(X_test)
```

```python
X_train=pd.DataFrame(X_train)
X_test=pd.DataFrame(X_test)
```

Train the model

```python
#  import all the regression model
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
# Evalution metries
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
# function of Evaluation metrix
def Evaluation(true,prediction):
    metrix_dic={
        "mean_squared_error":mean_squared_error(true,prediction),
        'mean_absolute_error':mean_absolute_error(true,prediction),
        'r2_score':r2_score(true,prediction),
        'Root mean square':np.sqrt(mean_squared_error(true,prediction))
    }
    return metrix_dic
```

```python
# create the object of the models
model={
    'Linear_Regression':LinearRegression(),
    'Lasso_regression':Lasso(),
    'Ridge_regression':Ridge(),
    'Knn_regression':KNeighborsRegressor(),
    'tree_regression':DecisionTreeRegressor(),
    'Random_forest_regression':RandomForestRegressor(),
    'Adaboost_regression':AdaBoostRegressor(),
    'Gradient_regression':GradientBoostingRegressor()
}

```

```python
# Make the prediction
for model_name,model_model in model.items():
    model_model.fit(X_train,y_train)

    # make the prediction;
    y_train_pred=model_model.predict(X_train)
    y_test_pred=model_model.predict(X_test)

    # evlauate the metrices
    print("Model name ---",model_name)
    print("Training Evlauation")
    print('\n')
    result=Evaluation(y_train,y_train_pred)
    for result_key,result_values in result.items():
        print(f"{result_key}:\n {result_values}")
    print("\n")
    print("Testing Evlauation")
    print('\n')
    result=Evaluation(y_test,y_test_pred)
    for result_key,result_values in result.items():
        print(f"{result_key}:{result_values}")
    print("-"*70)
```

Hyperparamter tuning

```python
# Linear regression
Linear_param_dist = {
    'fit_intercept': [True, False],
    'copy_X': [True, False],
    'positive': [True, False],
    'n_jobs': [-1, None]
}
# Lasso regression
Lasso_param_dist = {
    'alpha': [0.0001, 0.001, 0.01, 0.1, 1, 5, 10],
    'fit_intercept': [True, False],
    'max_iter': [500, 1000, 2000, 5000],
    'selection': ['cyclic', 'random'],
    'positive': [True, False]
}
# Ridge regression
Ridge_param_dist = {
    'alpha': [0.0001, 0.001, 0.01, 0.1, 1, 5, 10, 100],
    'fit_intercept': [True, False],
    'solver': ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga'],
    'max_iter': [None, 500, 1000, 2000],
    'positive': [True, False]
}

# knn regrerssion
Knn_param_dist = {
    'n_neighbors': [1, 2, 3, 4, 5, 7, 10, 15, 20, 30, 40, 50],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'p': [1, 2],
    'leaf_size': [10, 20, 30, 40, 50],
    'metric': ['minkowski', 'euclidean', 'manhattan']
}

# Decision tree regression

Tree_param_dist = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter': ['best', 'random'],
    'max_depth': [None, 5, 10, 20, 30, 50],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4, 6, 8],
    'max_features': [None, 'sqrt', 'log2']
}

# Random forest resgression
RandomForest_param_dist = {
    'n_estimators': [50, 100, 150, 200, 300, 400],
    'criterion': ['squared_error', 'absolute_error', 'friedman_mse', 'poisson'],
    'max_depth': [None, 5, 10, 20, 30, 50],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4, 6],
    'max_features': ['sqrt', 'log2', None],
    'bootstrap': [True, False],
    'n_jobs': [-1]
}
# Adaboost regression
Adaboost_param_dist = {
    'n_estimators': [50, 100, 150, 200, 300, 400, 500],
    'learning_rate': [0.01, 0.05, 0.1, 0.2, 0.5, 1.0, 1.5],
    'loss': ['linear', 'square', 'exponential']
}
# Gradient Boosting regression
Gradient_param_dist = {
    'n_estimators': [50, 100, 150, 200, 300, 400],
    'learning_rate': [0.01, 0.05, 0.1, 0.2, 0.3, 0.5, 1.0],
    'max_depth': [2, 3, 4, 5, 6],
    'min_samples_split': [2, 3, 5, 7, 10],
    'min_samples_leaf': [1, 2, 3, 4, 5],
    'subsample': [0.6, 0.7, 0.8, 0.9, 1.0],
    'max_features': ['sqrt', 'log2', None],
    'loss': ['squared_error', 'absolute_error', 'huber', 'quantile'],
    'criterion': ['friedman_mse', 'squared_error']
}
```

```python
# create the list of the paramter

hyper_paramter=[
    ("LinearRegression",LinearRegression(),Linear_param_dist),
    ("Lasso",Lasso(),Lasso_param_dist),
    ("Ridge",Ridge(),Ridge_param_dist),
    ('KNeighborsRegressor',KNeighborsRegressor(),Knn_param_dist),
    ('DecisionTreeRegressor',DecisionTreeRegressor(),Tree_param_dist),
    ("RandomForestRegressor",RandomForestRegressor(),RandomForest_param_dist),
    ("AdaBoostRegressor",AdaBoostRegressor(),Adaboost_param_dist),
    ("GradientBoostingRegressor",GradientBoostingRegressor(),Gradient_param_dist)
]
```

Hyperparamter tuning using randomsearchcv

```python
# randomSearchcv
from sklearn.model_selection import RandomizedSearchCV
model_param={}

for model_name1,model_model1,params in hyper_paramter:
    random=RandomizedSearchCV(estimator=model_model1,param_distributions=params,
                              cv=5,
                              verbose=2,
                              scoring='neg_root_mean_squared_error'
                              ,n_jobs=-1)
    
    # Train the model
    random.fit(X_train,y_train)

    # best param
    model_param[model_name1]=random.best_params_
    
```

```python
# print the best param
for Param_model_name,param_value in model_param.items():
    print(F'-------------------------Best Parameters for {Param_model_name}-------------')
    print(param_value)
```
